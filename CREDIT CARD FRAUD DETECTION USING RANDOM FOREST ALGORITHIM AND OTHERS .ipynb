# DATA PREPROCESSING 


```python
 from sklearn.preprocessing import StandardScaler  #  StandardScaler is used to resize the distribution of values ​​so that the mean of the observed values ​​is 0 and the standard deviation is 1  
```

```python
data['Amount']
```

```python
data['Amount'].values
```

```python
data['Amount'].values.reshape(-1,1)
```

```python
data["normalized data"] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
```

```python
data.drop(['Amount',"Time"],axis = 1).head()
```

```python
X = data.iloc[:,data.columns !="Class"]
y =data.iloc[:,data.columns =="Class"]  # use the double quotes to  define the coulumn 
```

```python
print(X.head())
```

```python
y.head()
```

```python
from sklearn.model_selection import  train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, random_state = 0,test_size = 0.33)
```

```python
X_train.shape
```

```python
X_test.shape
```

# MODEL 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest = RandomForestClassifier(n_estimators = 100) #number of trees
```

```python
randomforest.fit(X_train,y_train.values.ravel())
```

```python
y_pred = randomforest.predict(X_test)
```

```python
print(y_pred)
```

```python
randomforest.score(X_test,y_test)
```

```python
cmatrix = confusion_matrix(y_test,y_pred)
print(cmatrix)
```

```python
labels = ["normal", "fraud"]
sns.heatmap(cmatrix,fmt = ".3f",annot = True,cmap = "YlGnBu",xticklabels = labels,yticklabels = labels)
plt.show()
```

# USING NAIVE BAYES

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model = BernoulliNB()
```

```python
model.fit(X_train,y_train)
```

```python
Y_pred = model.predict(X_test)
```
